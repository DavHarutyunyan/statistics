##Task 1: Calculate the confidence intervals for men shoes sales in the USA, this time based on a bigger sample - 2015-2016. 

```python
start_date_15 = '2015-01-01'
end_date_15 = '2015-12-31'

start_date_formatted_15 = pd.to_datetime(start_date_15)
end_date_formatted_15 = pd.to_datetime(end_date_15)

gender_mask = df['Gender'] == 'Male'
date_mask_15 = (df['Date'] >= start_date_formatted_15) & (df['Date'] <= end_date_formatted_15)

combined_mask_15 = gender_mask & date_mask_15
men_shoes_sales_15 = df.loc[combined_mask_15]

men_shoes_sales_15
```

```python
start_date_16 = '2016-01-01'
end_date_16 = '2016-12-31'

start_date_formatted_16 = pd.to_datetime(start_date_16)
end_date_formatted_16 = pd.to_datetime(end_date_16)

gender_mask = df['Gender'] == 'Male'
date_mask_16 = (df['Date'] >= start_date_formatted_16) & (df['Date'] <= end_date_formatted_16)

combined_mask_16 = gender_mask & date_mask_16
men_shoes_sales_16 = df.loc[combined_mask_16]

men_shoes_sales_16
```

```python
pivot_table_15 = men_shoes_sales_15.pivot_table(index='Size (US)', columns= men_shoes_sales_15['Date'].dt.month, aggfunc='size', fill_value=0)
```

```python
pivot_table_16 = men_shoes_sales_16.pivot_table(index='Size (US)', columns= men_shoes_sales_16['Date'].dt.month, aggfunc='size', fill_value=0)
```

```python
pivot_table_15.reset_index(drop=True, inplace=True)
```

```python
pivot_table_15['Mean'] = [round((np.sum(pivot_table_15.iloc[i,:])/len(pivot_table_15)),2) for i in range(len(pivot_table_15))]
```

```python
pivot_table_15['Variance'] = [round((np.sum((pivot_table_15.iloc[i, :] - pivot_table_15['Mean'][i])**2)/11),2) for i in range(len(pivot_table_15))]
```

```python
pivot_table_15['std'] = [round((pow((x),1/2)),2) for x in pivot_table_15['Variance']]
```

```python
pivot_table_15['Standard error'] = [round((x/pow((len(pivot_table_15)),1/2)),2) for x in pivot_table_15['std']]
```

```python
df_z_table = pd.read_csv('z_table.csv', sep = '\t')
```

```python
df_z_table
```

```python
def z_statistic(confidence_percent, table):
    alpha_half = (100 - confidence_percent) / 200  # Adjusted alpha value
    z_value = 1 - alpha_half
    indices, columns = np.where(np.isclose(table.values, z_value, atol=0.001))

    if len(indices) > 0:
        return round((table['z'][indices[0]] + float(table.columns[columns[0]])),2)
    else:
        print('z_statistic not calculated')
        return None
```

```python
z_statistic(95, df_z_table)
```

```python
pivot_table_15['Margin of error'] = [round((x * z_statistic(95, df_z_table)),2) for x in pivot_table_15['Standard error']]
```

```python
pivot_table_15['95% Confidence level lower'] = [pivot_table_15['Mean'][i] - pivot_table_15['Margin of error'][i] for i in range(len(pivot_table_15))]
pivot_table_15['95% Confidence level upper'] = [pivot_table_15['Mean'][i] + pivot_table_15['Margin of error'][i] for i in range(len(pivot_table_15))]
```

```python
pivot_table_15.insert(loc=0, column = 'Size (US)', value = np.sort(men_shoes_sales_15['Size (US)'].unique()))
```

```python
pivot_table_15 = pivot_table_15.reset_index(drop=True)
pivot_table_15
```
