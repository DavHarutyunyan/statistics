```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('reading_writting.csv', sep='\t')
x = df['Writing']
y = df['Reading']
df
```

```python
def mean(ds):
    return sum(ds)/len(ds)    
```

```python
def covariance(ds1, ds2):
    return sum((ds1-mean(ds1))*(ds2-mean(ds2)))/(len(ds1)-1)
```

```python
print('sample of 5 students covariance = ', covariance(x,y))
```
